```python
import pandas as pd
import numpy as np
import cv2
import efficientnet.tfkeras as efn
from keras.models import load_model
import warnings
warnings.filterwarnings("ignore")
#load the model
def load_oral_model(model):
    model = load_model(model)
    return model
    
def pred_result(img_path, model):
    img = cv2.imread(img_path)
    img1 = cv2.resize(img, (224,224))
    # convert the image to a tensor for inference
    img1 = np.expand_dims(img1, 0).astype(np.float32) / 255.0
    model = load_oral_model(model)
    preds = np.squeeze(model.predict(img1)[0])
    #preds will give the probability score of 
    class_name = ['healthy_nails','healthy_skin','unhealthy_nails','unhealthy_skin']
    index = np.argmax(preds)
    max_prob = np.amax(preds)
    return img,preds, max_prob, class_name[index]


img_path = "E:/Bhuwan/malnutrition/Datasets/images/test/unhealthy_skin/5.jpg" 
file_name = img_path.split('/')[-1]
filename = file_name.split('.')[0]
print("file_name: ",file_name)
model_path = "E:/Bhuwan/malnutrition/malnutrition.h5"
dest_path = "output/"
img,preds,max_prob, class_name = pred_result(img_path,model_path)
print("prediction:",preds) 
print("highest probability:",max_prob)
if class_name == "healthy_nails":
    
    cv2.putText(img, class_name, (50,100), cv2.FONT_HERSHEY_SIMPLEX, 
                   1, (0, 0, 255), 3, cv2.LINE_AA)
    cv2.imwrite(dest_path+file_name,img)
    cv2.imshow(f"{filename}",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("The uploaded image belongs to ",class_name ,"class.","and the person don't have to worry. Hence He has a healthy nails")

elif class_name == "healthy_skin":
    cv2.putText(img, class_name, (50,100), cv2.FONT_HERSHEY_SIMPLEX, 
                   1, (0, 0, 255), 3, cv2.LINE_AA)
    cv2.imwrite(dest_path+file_name,img)
    cv2.imshow(f"{filename}",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("The uploaded image belongs to ",class_name ,"class.","and the person don't have to worry. Hence He has a healthy skin")
   
elif class_name == "unhealthy_nails":
    cv2.putText(img, class_name, (50,100), cv2.FONT_HERSHEY_SIMPLEX, 
                   1, (0, 0, 255), 3, cv2.LINE_AA)
    cv2.imwrite(dest_path+file_name,img)
    cv2.imshow(f"{filename}",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("The uploaded image belongs to ",class_name ,"class.","and the person have to clean theirs nails. Hence Report is Negative") 
   
elif class_name == "unhealthy_skin":
    cv2.putText(img, class_name, (50,100), cv2.FONT_HERSHEY_SIMPLEX, 
                   1, (0, 0, 255), 3, cv2.LINE_AA)
    cv2.imwrite(dest_path+file_name,img)
    cv2.imshow(f"{filename}",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    print("The uploaded image belongs to ",class_name ,"class.","and the person need to visit the doctor. Hence Report is positive")
    
```
